# Gaussian leakage

Generated by the convolution of the Sinc() with the transfer function of the Gaussian kernel.

```python
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
from scipy import signal
from denoising.utils import get_gaussian_kernel

# Define parameters
beta = 0.5 # 1/np.sqrt(2)
tau = 5  # Range of sigma values to test
#sigma_values = np.linspace(0.5, 0.5, 1)  # Range of sigma values to test
bandwidth=1
normalized_cutoff_freqs = []

def naive_gaussian_kernel(sigma=1, k=4):
    """Generates a truncated Gaussian kernel without smoothing adjustments."""
    L = 2 * k * sigma + 1  # Ensure an odd length
    x = np.arange(-L//2, L//2 + 1)
    kernel = np.exp(-0.5 * (x / sigma) ** 2)
    return kernel / kernel.sum()  # Normalize

kernel = naive_gaussian_kernel(5, 1)
freqs, response = signal.freqz(kernel, fs=bandwidth)
response = np.abs(response)
plt.plot(freqs, 20 * np.log10(response), label="$k=1$")

kernel = naive_gaussian_kernel(5, 4)
freqs, response = signal.freqz(kernel, fs=bandwidth)
response = np.abs(response)
plt.plot(freqs, 20 * np.log10(response), label="$k=4$")

plt.xlabel(r"$f/f_s$")
plt.ylabel(r"$20\log|\mathcal{F}(\mathbf{h})(f)|$ [dB]")
plt.title(rf"$\tau={tau}$")
plt.legend()
#plt.plot(freqs, response)
plt.savefig("gaussian_leakage.pdf", bbox_inches='tight')
```
